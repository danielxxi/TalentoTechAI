```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Definir el número de filas para cada archivo
num_filas = 100

# Generar datos aleatorios para el primer archivo
data1 = pd.DataFrame({
    'ID': range(1, num_filas + 1),
    'Columna_A': np.random.randint(0, 100, num_filas),
    'Columna_B': np.random.rand(num_filas),
    'Columna_C': np.random.randint(0, 100, num_filas)
})

# Generar datos aleatorios para el segundo archivo
data2 = pd.DataFrame({
    #'ID': range(1, num_filas + 1),
    'ID': range(101, 101 + num_filas),
    'Columna_D': np.random.randint(50, 150, num_filas),
    'Columna_E': np.random.uniform(1, 10, num_filas),
    'Columna_F': np.random.rand(num_filas)
})

# Guardar los datos en archivos Excel
data1.to_excel('archivo1.xlsx', index=False)
data2.to_excel('archivo2.xlsx', index=False)


# Concatenar los DataFrames uno debajo del otro
data_combined = pd.concat([data1, data2])

print("_______________________________________________________________________")
data_combined.columns
print(data_combined.columns)

# Combinar los DataFrames, por ejemplo, fusionándolos por una columna común
# Si deseas fusionarlos por una columna específica, puedes usar la función merge de Pandas
# Por ejemplo, si tienen una columna 'ID' común, puedes fusionarlos así:
# data_merged = pd.merge(data1, data2, on='ID')
# Fusionar los DataFrames por la columna 'ID'
data_merged = pd.merge(data1, data2, on='ID')

print("_______________________________________________________________________")
print("_______________________________________________________________________")

# Realizar el análisis deseado utilizando los DataFrames combinados
# Por ejemplo, puedes calcular estadísticas descriptivas, correlaciones, etc.
print("Estadísticas descriptivas del DataFrame combinado:")
print(data_combined.describe())
print("_______________________________________________________________________")
print("\nCorrelación entre las columnas del DataFrame combinado:")
print(data_combined.corr())

# Generar visualizaciones y gráficos para explorar los datos
# Por ejemplo, puedes crear gráficos de dispersión, histogramas, etc.
data_combined.plot.scatter(x='Columna_A', y='Columna_C')
print("_______________________________________________________________________")
data_combined['Columna_D'].plot.hist()

print("_______________________________________________________________________")
print("\n DataFrame combinado:")
# Mostrar los resultados 
print(data_combined.head(200))
print("_______________________________________________________________________")
#o guardarlos en otro archivo si es necesario
data_combined.to_excel('archivo_combinado.xlsx', index=False)
print("_______________________________________________________________________")
# Mensaje de confirmación
print("Los archivos Excel se han generado y combinado correctamente.")
print("_______________________________________________________________________")
print("_______________________________________________________________________")
print("_______________________________________________________________________")
print("\n DataFrame fusionado:")
# Mostrar los resultados 
print(data_merged.head(200))
data_merged.to_excel('archivo_fusionado.xlsx', index=False)
# Mensaje de confirmación
print("Los archivos Excel se han generado y fusionado correctamente.")
print("_______________________________________________________________________")
```

# Precisiones y consideraciones comunes


Coincidencia de índices: Si ambos archivos tienen un campo de índice común o una columna que puede servir como identificador único para cada fila, puedes fusionar los datos en función de este campo utilizando pandas.

Dimensiones similares: Si los archivos tienen diferentes dimensiones, puedes optar por fusionarlos verticalmente (apilando filas) o horizontalmente (concatenando columnas), dependiendo de cómo quieras combinar los datos.

Campos comunes: Si ambos archivos tienen columnas con nombres similares o datos que pueden relacionarse de alguna manera, puedes fusionarlos en función de estas columnas utilizando métodos como merge() o concat() de pandas.

La forma en que se combinen los archivos dependerá de la estructura de los datos y de lo que desees lograr con la fusión o concatenación.  

No necesariamente los dos archivos deben tener las mismas dimensiones o contener el mismo campo índice. La fusión o concatenación de archivos Excel generalmente se realiza en función de alguna relación entre los datos que se desea combinar.
