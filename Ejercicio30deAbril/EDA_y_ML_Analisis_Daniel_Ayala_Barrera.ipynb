## Ahora el problema dado es un problema de clasificación.. Así que vamos a utilizar los algoritmos de clasificación para construir un modelo.

### Clasificación: Las muestras pertenecen a dos o más clases y queremos aprender de los datos ya etiquetados cómo predecir la clase de los datos no etiquetados

### Regresión: Si el resultado deseado consiste en una o más variables continuas, la tarea se denomina regresión. Un ejemplo de problema de regresión sería la predicción de la longitud de un salmón en función de su edad y peso.

```python
# importando todos los paquetes necesarios para usar los distintos algoritmos de clasificación
from sklearn.linear_model import LogisticRegression # para el algoritmo de regresión logística
from sklearn.model_selection import train_test_split # para dividir el conjunto de datos para entrenamiento y prueba
from sklearn.neighbors import KNeighborsClassifier # clasificador KNN
from sklearn import svm # para el algoritmo support vector machine
from sklearn import metrics # para comprobar la precisión del modelo
from sklearn.tree import DecisionTreeClassifier # para usar DTA
```

Ahora bien, cuando entrenamos cualquier algoritmo, el número de características y su correlación desempeñan un papel importante. Si hay características y muchas de ellas están muy correlacionadas, el entrenamiento de un algoritmo con todas las características reducirá la precisión. Por lo tanto, la selección de características debe hacerse con cuidado.

```python
# Check for non-numeric values in each column
saber_agreg_df.columns
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
saber_agreg_df['CALENDARIO']= label_encoder.fit_transform(saber_agreg_df['CALENDARIO'])

saber_agreg_df['CALENDARIO'].unique()
```
